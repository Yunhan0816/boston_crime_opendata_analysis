```python
larceny = df[df.OFFENSE_CODE_GROUP.str.contains("Larceny")]
larceny.head()
```

```python
larceny.shape
```

```python
larceny.groupby("STREET").size().sort_values(ascending = False).head(10).sort_values().plot(kind = "barh")
```

```python
larceny.groupby("MONTH").size().sort_values().plot(kind = "bar")
```
